```python
import pandas as pd
```

```python
people = {
    'first': ['Corey', 'Jane', 'John'], 
    'last': ['Schafer', 'Doe', 'Doe'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['first'] + ' ' + df['last'] 
```

```python
df['FullName'] = df['first'] + ' ' + df['last'] #we can use map also
```

```python
df
```

```python
df.drop(columns=['first','last'], inplace=True)
```

```python
df
```

```python
df['FullName'].str.split(' ')
```

```python
df['FullName'].str.split(' ',expand = True)
```

```python
df[['first', 'last']] = df['FullName'].str.split(' ',expand = True)
```

```python
df
```

```python
df.concat({'first' : 'tony'}, ignore_index = True)  
```

```python
import pandas as pd

# 1. Create a temporary DataFrame for your new row
new_row = pd.DataFrame([{'first': 'tony'}])

# 2. Use pd.concat() instead of df.concat()
df = pd.concat([df, new_row], ignore_index=True)
```
